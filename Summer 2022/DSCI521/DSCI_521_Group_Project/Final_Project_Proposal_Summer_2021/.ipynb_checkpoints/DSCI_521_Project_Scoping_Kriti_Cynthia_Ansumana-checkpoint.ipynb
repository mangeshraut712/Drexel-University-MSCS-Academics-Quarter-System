# Project Objectives

For this project, we perform exploratory data analysis on a spotify dataset to summerize their main characteristics, with visual methods.

**Data** 

We are using an existing dataset present on kaggel. The dataset consist of audio features of approximately 600k songs released in between 1922 and 2021, along with the data of 1.1 million artists.

Includes: <br>
1) Tracks metadata <br>
2) Audio features <br>
3) Artists attribute <br>



#Exhibition of Analysis



**Description of the dataframes** <Br>
The 2 dataframes we have begun to explore are tracks.csv and data_by_genre.csv. We loaded them and assigned them to the variables tracks and genres respectively. <br>
<Br>
Tracks and genres both contain the following 13 variables:
- acousticness
- danceability
- duration
- energy
- instrumentalness
- key
- liveness
- loudness
- mode
- popularity
- speechiness
- tempo
- valence

Genre also includes an identifying variable:
- genres

Tracks includes an additional 4 variables:
- explicit
- id
- name
- time signature


The true way that Spotify collects and classifies audio feature data is undisclosed. The descriptions of these measurements are company property. Our project seeks to further understands the relationships between these numeric variables and their meanings.

Tracks records data for these variables for 586601 songs.
Genre records the average of the observations for all tracks in that genre.
