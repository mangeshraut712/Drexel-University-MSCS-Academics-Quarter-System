```python
import pandas as pd

data = pd.read_csv('./tweet_emotions.csv', delimiter=',')
```

```python
data["sentiment"].value_counts()
```

```python
import re
# Filter out nonsense words from list of words (punctuation, usernames, URLs)
def preprocess(content):
  content = re.sub(r'http\S+','', content)
  content = re.sub(f'@\S+','', content)
  content = re.sub(r'[^\w\s]', '', content)
  cleaned_content = []
  for i in content.split():
    if i.strip().lower().isalpha():
      cleaned_content.append(i.strip().lower())
  return cleaned_content

data['content'] = data['content'].apply(preprocess)
```

```python
from collections import Counter
from collections import defaultdict

counter = {}
for sentiment in data['sentiment']:
  counter[sentiment] = Counter()
```

```python
#sentiment analysis
from textblob import TextBlob
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        result = "Positive"
    elif sentiment < 0:
        result = "Negative"
    else:
        result = "Neutral"
    return result
        
```

```python
# Text function
get_sentiment("I love coding")
```

```python
from collections import Counter
```

```python
def extract_keywords(text,num=50):
    tokens = [ tok for tok in text.split()]
    most_common_tokens = Counter(tokens).most_common(num)
    return dict(most_common_tokens)
```

```python
#list of emotions
emotion_list = df['sentiment'].unique().tolist()
```

```python
emotion_list
```

```python
love_list = df[df['sentiment'] == 'love']['Clean_Text'].tolist()
```

```python
#love document
love_docx = ' '.join(love_list)
```

```python
love_docx
```

```python
#extract keywords
keyword_love = extract_keywords(love_docx)
```

```python
keyword_love
```

```python
#plot
def plot_most_common_words(mydict,emotion_name):
    df_01 = pd.DataFrame(mydict.items(),columns=['tokens','count'])
    plt.figure(figsize=(20,10))
    plt.title("Plot of {} Most common keywords.".format(emotion_name))
    sns.barplot(x='tokens',y='count',data=df_01)
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_most_common_words(keyword_love,"love")
```

```python
sadness_list = df[df['sentiment']== 'sadness']['Clean_Text'].tolist()
sadness_docx = ' '.join(sadness_list)
keyword_sadness = extract_keywords(sadness_docx)
```

```python
plot_most_common_words(keyword_sadness,'sadness')
```

```python
#load ML packages
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

#vectorizer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

#metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
#split our dataset
from sklearn.model_selection import train_test_split
```

```python
#build features from our data
Xfeatures = df['Clean_Text']
ylabels = df['sentiment']
```

```python
Xfeatures
```

```python
#Vectorizer
cv = CountVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
#get features by name 
cv.get_feature_names()
```

```python
#to dense array(Numpy)
X.toarray
```

```python
#split dataset
X_train,X_test,y_train,y_test = train_test_split(X,ylabels,test_size=0.5,random_state=45)
```

```python
#build our model
nv_model = MultinomialNB()
nv_model.fit(X_train,y_train)
```

```python
#accuracy
nv_model.score(X_test,y_test)
```

```python
#predictions
y_pred_for_nv = nv_model.predict(X_test)
```

```python
y_pred_for_nv
```

```python
#make single prediction
    #vectorized our model
    #applied our model
```

```python
sample_text = ["I love coding so much"]
```

```python
vect = cv.transform(sample_text).toarray()
```

```python
#make prediction
nv_model.predict(vect)
```

```python
#check for prediction probability (percentage)/confidence score
nv_model.predict_proba(vect)
```

```python
#get all class for our model
nv_model.classes_
```

```python
np.max(nv_model.predict_proba(vect))
```

```python
def predict_emotion(smaple_text,model):
    myvect = cv.transform(sample_text).toarray()
    prediction = model.predict(myvect)
    pred_prob = model.predict_proba(myvect)
    pred_percentage_for_all = dict(zip(model.classes_,pred_prob[0]))
    print("Prediction:{},Prediction Score:{}".format(prediction[0],np.max(pred_prob)))
    print(prediction[0])
    return pred_percentage_for_all
```

```python
predict_emotion(sample_text,nv_model)
```

```python
predict_emotion(["He hates running all day"],nv_model)
```

```python
#model classification
print(classification_report(y_test,y_pred_for_nv))
```

```python
#confusion_matrix
confusion_matrix(y_test,y_pred_for_nv)
```
